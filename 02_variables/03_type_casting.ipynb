# 1. Type Casting in Python

## int to float

```python
age_int = 25 # this is an integer
age_float = float(age_int)
```

```python
print(type(age_int))
print(type(age_float))
```

## float to int

```python
num_float = 28.96
num_int = int(num_float)
```

```python
print(type(num_int))
```

```python
price = 19.99
price_int = int(price)
```

```python
print(type(price_int))
print(price_int)
```

## float or int to string

```python
number_int = 15
number_float = 16.8
```

```python
number_int = str(number_int)
number_float = str(number_float)
```

```python
type(number_int)
```

```python
type(number_float)
```
