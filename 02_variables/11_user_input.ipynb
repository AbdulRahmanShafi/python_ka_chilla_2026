# Validation Input 

```python
age = input("What is your age? ")
if age.isdigit():
    age = int(age)
    print("Your age is:", age)
else:
    print("Invalid input. Write age in numeric form.")
```

```python
name = input("What is your name? ")
print(f"Hello, {name}!")
```
