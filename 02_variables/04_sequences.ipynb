# Sequence in python

1. lenght of sequence

```python
my_list = [1, 2, 3, 4, 5, 6, 7]
print(len(my_list))
```

2. Membership test

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print("apple" in fruits)
```

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print("mango" in fruits)
```

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print("mango" not in fruits)
```

3. Concatenation

```python
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
combined = list1 + list2
print(combined)
```

4. Repetition

```python
my_list = [1, 2, 3, 4, 5]
print(my_list * 2)
```
