## Lambda Function for addition

```python
# Lambda function
# syntax: lambda arguments : expression
add_lambda = lambda a, b : a + b
```

```python
add_lambda(89, 37)
```

### For Subtraction

```python
subtract_lambda = lambda a, b, c: a - b - c
subtract_lambda(89, 37, 12)
```

### For Multiply

```python
square = lambda x: x * x
square(79)
```

```python
multiply_lambda = lambda a, b: a * b
multiply_lambda(89, 37)
```

### Even or Odd check

```python
check = lambda x: "Even" if x % 2 == 0 else "Odd"
check(7)
```

```python
check = lambda x: "Even" if x % 2 == 0 else "Odd"
check(6)
```

### For division

```python
divide_lambda = lambda a, b: a / b 
divide_lambda(89, 37)
```

### List sorting 

```python
marks = [50, 30, 90, 70]
marks.sort(key=lambda x: x)
marks
```

```python
number = [5, 20, 15, 18, 30, 35, 45, 56]
# lambda function
squared_numbers = list(map(lambda x: x**2, number))
```

```python
squared_numbers
```

```python
number = [5, 20, 15, 18, 30, 35, 45, 56]
# lambda function
cube_numbers = list(map(lambda x: x**3, number))
```
