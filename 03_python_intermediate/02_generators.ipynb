# Generators in python

```python
# generators functions
def my_gen():
    yield 'A'
    yield 'B'
    yield 'C'
    yield 'D'
    
# generator object
g = my_gen()

# print generator object
print(next(g))
```

```python
print(next(g))
```

```python
# Gnerators expression
gen = (x**2 for x in range(5))
for x in gen:
    print(x)
```
