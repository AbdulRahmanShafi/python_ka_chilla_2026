```python
print("My name is Abdul Rahman")
print("Hello World!")
print("Hello, World!")
print("Hello" , "World!")
print("Hello", "World!", sep=", ")
print("Hello", "Abdul Rahman!", sep=": ", end="\n")
print("Welcome", "Abdul Rahman ", sep=": ", end="Shafi\n")
```

## f-string

```python
# f-string is a way to format strings in Python. 
name = "Abdul Rahman"
age = 20
final = "How are you today?"
print(f"My name is {name}, and I am {age} years old. {final}")
```

```python
# f-string with line break
name = "Abdul Rahman"
age = 20
final = "How are you today?"
print(f"My name is {name}, and I am {age} years old.\n{final}")
# \n break the line and move to the next line.
```

```python
# str.format()
print("Hello, {}!".format("Codanics"))
```

```python
print("Hello, {}:".format("Abdul Rahman"))
```

```python
## % formatting
name = "Abdul Rahman"
age = 20
print("My name is %s and I am %d years old." %(name, age))
# %s for string and %d for integer. We can also use %f for float and %c for character.
```

```python
name = "Hamza"
age = 25
section = "Shaheen"
print("My name is %s : I am %d years old and I am in section %s." %(name, age, section))
# %s for string and %d for integer. We can also use %f for float and %c for character.
```

```python
name = "Abdul Rahman"
age = 20.5
print("My name is %s and I am %f years old." %(name, age))
# %s for string and %d for integer. We can also use %f for float and %c for character.
```
