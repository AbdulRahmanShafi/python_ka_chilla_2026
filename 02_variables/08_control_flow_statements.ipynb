# **_Logical Operators in if else conditions_**

## Use of AND Operation with if else condition

```python
age = 25
citizen = True
if age >= 18 and citizen:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
age = 17
citizen = True
if age >= 18 and citizen:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

## Use of OR Operator with if else condition

```python
day = "Sunday"
if day == "Saturday" or day == "Sunday":
    print("It's the weekend! Time to relax.")
else:
    print("It's a weekday. Time to work!")
```

```python
day = "Tuesday"
if day == "Saturday" or day == "Sunday":
    print("It's the weekend! Time to relax.")
else:
    print("It's a weekday. Time to work!")
```

## Use of NOT Operator with if else condition

```python
is_raining = False
if not is_raining:
    print("It's a nice day! You can go outside.")
else:
    print("It's raining. Better stay indoors.")
```

```python
is_raining = True
if not is_raining:
    print("It's a nice day! You can go outside.")
else:
    print("It's raining. Better stay indoors.")
```

```python
age = 18
license = True
if age >= 18:
    if license:
        print("You can drive.")
    else:
        print("You need a license to drive.")
else:
    print("You are too young to drive.")
```

## Break and Continue Statement

```python
cups_of_chai = 0

while cups_of_chai < 5:
    cups_of_chai += 1
    print("I want more chai.")
    if cups_of_chai == 3:
        print("Too much chai, stopping now.")
        break
    print("I have had", cups_of_chai, "cups of chai.")
```

```python
cups_of_chai = 0

while cups_of_chai < 5:
    cups_of_chai += 1
    print("I want more chai.")
    if cups_of_chai == 3:
        print("Too much chai, stopping now.")
        break
    print("I have had", cups_of_chai, "cups of chai.")
```

## continue statement

```python
cups_of_chai = 0

while cups_of_chai < 5:
    cups_of_chai += 1
    print("I want more chai.")
    if cups_of_chai == 2:
        print("Too much chai, stopping now.")
        continue
    print("I have had", cups_of_chai, "cups of chai.")
```

# Pass Statement

```python
cups_of_chai = 0
while cups_of_chai < 5:
    cups_of_chai += 1
    print("I want more chai.")
    if cups_of_chai == 3:
        print("Too much chai, stopping now.")
        continue # You can use pass to do nothing and continue with the next iteration of the loop.
    print("I have had", cups_of_chai, "cups of chai.")
```

```python
cups_of_chai = 0
while cups_of_chai < 5:
    cups_of_chai += 1
    print("I want more chai.")
    if cups_of_chai == 3:
        print("Too much chai, stopping now.")
        pass 
    print("I have had", cups_of_chai, "cups of chai.")
```

## try, except and finally

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
finally:
    print("This will always be executed.")
```

### Real World Example

try     → paisay nikalnay ki koshish
except  → agar card block / balance kam ho
finally → card wapas nikalna (HAR HAAL MEIN)

```python
try:
    File = open("report.txt", "r")
    Data = File.read()
except FileNotFoundError:
    print("File nhi mili.")
finally:
    print("File handling is done.")
```

```python
try:
    print("Hello, World!")
    print(10 / 0) # This will raise an exception
except ZeroDivisionError:
    print("You cannot divide by zero!")
finally:
    print("This will always be executed.")
```
