### Defining Mean Function

```python
def mean(numbers):
    """This function will calculate the mean of the numbers provided as an argument."""
    return sum(numbers)/len(numbers)
```

```python
prices = [100, 200, 300, 400, 650, 750, 850, 900, 985, 990, 995,]
mean(prices)
```

```python
def mean(numbers):
    """This function will calculate the mean of the numbers provided as an argument."""
    return round(sum(numbers)/len(numbers), 2)
```

```python
prices = [100, 200, 300, 400, 650, 750, 850, 900]
mean(prices)
```

```python
def add(a, b, c):
    """This function will add three numbers provided as arguments."""
    return a + b + c
```

```python
add(256, 512, 1024)
```

```python
def introduce(name, age, city):
    """This function will introduce a person with the name, age, and city provided as arguments."""
    print("Hello, My name is " + name + ". I am " + str(age) + " years old and I live in " + city + ".")
```

```python
introduce("Abdul Rahman", 20, "Lahore")
```

### For Addition

```python
def add(a, b):
    """This function will add two numbers provided as arguments."""
    return a + b
add(124, 236)
```

```python
def subtract(a, b):
    """This function will subtract two numbers provided as arguments."""
    return a - b
subtract(986, 248)
```

```python
def multiply(a, b):
    """This function will multiply two numbers provided as arguments."""
    return a * b
multiply(568, 1498)
```

```python
def divide(a, b):
    """This function will divide two numbers provided as arguments."""
    return a / b
divide(457, 23)
```
