# ***Mutable Elements***

```python
fruits = ['apple', 'banana', 'orange']
```

```python
fruits[1] = 'grape'
print(fruits)
```

```python
list = ["Ali", "Ahmed", "Sara", "Lina", "Omar", "Nour", "Yousef", "Mona", "Huda", "Khaled"]
list[7] = "Rana"
```

```python
print(list)
```

# Tuple

```python
tuple = ('apple', 'banana', 'orange')
tuple[1] = 'grape'  # This will raise a TypeError because tuples are immutable
```
