# _Variables_
Variable is a name that refers to a value. The value can be changed, hence the name variable.In Python, the variables can be declared and values can be assigned to it as follow:
These are the types of variables in python:

1. Numbers
2. String
3. List
4. Tuple
5. Dictionary
6. Set
7. Boolean


```python
age = 20
```

```python
print(age)
```

```python
print("I am", age, "Years old.")
```

```python
name = "My name is Abdul Rehman."
```

```python
print(name, "I am", age, "Years old.")
```
