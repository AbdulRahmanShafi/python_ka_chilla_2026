# **_Types of Variables_**

# 1. Integer
_Stores whole numbers without decimals._

```python
age = 10
type(age)
print(type(age))
```

# 2. Floating
_Stores numbers with decimal points._

```python
temperature = 12.5
```

```python
type(temperature)
print(type(temperature))
```

# 3. String
_Stores text or characters enclosed in quotes._

```python
greetings = "Hello, Codanians!"
```

```python
type(greetings)
```

```python
print(type(greetings))
```

# 4. Boolean
_Stores logical values: True or False._

```python
is_raining = True
```

```python
type(is_raining)
```

```python
print(type(is_raining))
```

# 5. List
_Stores multiple items in one variable; ordered and changeable._

```python
numbers = [1,2,3,4,5]
type(numbers)
print(type(numbers))

names = ["Ali", "Hamza", "Sahar"]
print(type(names))
```

```python
print(names)
print(numbers)
```

# 6. Tuple
_Stores multiple items; ordered and unchangeable._

```python
prices = (10.2, 12.5, 14.6, 11.0)
print(type(prices))
type(prices)
```

# 7. Dictionary
_Stores data in key–value pairs._

```python
student = {"name": "Ayesha", "age": 20, "courses": ["Math", "Science"]}
print(type(student))
print(student)
```

# 8. Set
_Sets are unordered collections of unique items, defined using curly braces {} or set() functions._

```python
set = {1, 2, 3, 4, 5}
print(type(set))
```

# 9. None Type
_None represents the absance of a value and is an object of its own datatype._

```python
result = None
print(type(result))
```

# 10. Complex
_Complex numbers are numbers with a real and imaginary part. The Imaginary part is denoted by j._

```python
complex = 3 + 4j
print(type(complex))
type(complex)
```
